```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

import requests
import time
import urllib.request
import os
import socket

# 디렉토리가 없는 경우 자동으로 생성해줌
def createDirectory(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print("Error: Failed to create the directory.")
        

def crawling_img(name):
    
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    driver.set_page_load_timeout(10)
    driver.get("https://www.google.co.kr/imghp?hl=ko&tab=wi&authuser=0&ogbl")
    elem = driver.find_element(By.NAME, "q")
    elem.send_keys(name)
    elem.send_keys(Keys.RETURN)
    
    # 시간을 넉넉히 3초로했다 (1초로 하니 스크롤이 먼저 끝나는 경우가 있음)
    SCROLL_PAUSE_TIME = 3
    # 브라우저 높이 초깃값
    last_height = driver.execute_script("return document.body.scrollHeight") 
    while True:
        
        # 브라우저 끝까지 스크롤을 내림
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  
        # 시간차를 둔다
        time.sleep(SCROLL_PAUSE_TIME)
        
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            try:
                # 결과 더보기 버튼
                driver.find_element(By.CSS_SELECTOR,".mye4qd").click()
            except:
                # 버튼이 없는 경우 종료
                break
        last_height = new_height
    # 작은 이미지들 모두 선택하고 imgs에 저장
    imgs = driver.find_elements(By.CSS_SELECTOR,".rg_i.Q4LuWd")
    # 디렉토리를 설정
    dir = "./car/" + name
    
    # 수집된 작은 이미지 길이 출력 (확인용)
    print(len(imgs))
    
    # 폴더 생성
    createDirectory(dir)
    
    # 파일 이름을 구분지어 주기위해 count 변수에 숫자를 넣어줌
    count = 1
    
    # 만들어진 경로를 저장
    path = "./car/" + name + "/"
    for img in imgs:
        
        try:
            # 15초 안에 아래 코드들이 실행되지 않으면 예외를 발생 시킴
            socket.setdefaulttimeout(15)
            
            # 작은 이미지 클릭
            img.click()
            time.sleep(2)
            
            # 큰 이미지의 주소를 imgUrl 변수에 저장
            imgUrl = driver.find_element(By.XPATH, '/html/body/div[2]/c-wiz/div[3]/div[2]/div[3]/div/div/div[2]/div[2]/div[2]/c-wiz/div[2]/div[1]/div[1]/div[2]/div/a/img').get_attribute("src")

            # url을 이용해 파일을 저장해주는 메서드
            urllib.request.urlretrieve(imgUrl, path + name + str(count) + ".jpg")
            
            # 진행과정이 잘되고 있는지 확인용
            print(path + name + str(count) + ".jpg", "저장 완료")
            print("operate normally");print("=====================================")
            
            # 위의 과정이 문제없이 됐을때 count에 1을 더해줌
            count += 1
            
            if count >= 600:
                break
        except:
            print('error')
            pass
    # 브라우저창 종료
    driver.close()
    
    
```

```python
arr = ['yellow sedan car', 'green compact car', 'green sedan car']
for i in arr:
    crawling_img(i)
```
